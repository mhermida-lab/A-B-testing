 An interesting finding is that not all users where assigned to the test the day it started. Every day after the first one a new batch of users were assigned to the test. In my oppinion this could lead to misleading insights as these users won't go through the whole process.   
I'm going to check if this phenomenon is present evenly in the two test groups. I that case I'll narrow my analysis to only users assigned to the test on the first day.

```python
df1.assignment_date.value_counts()
```

```python
groupA.assignment_date.value_counts()
```

```python
groupB.assignment_date.value_counts()
```

```python
(df1.assignment_date.value_counts()/df1.assignment_date.value_counts().sum())*100
```

```python
(groupA.assignment_date.value_counts()/groupA.assignment_date.value_counts().sum())*100
```

```python
(groupB.assignment_date.value_counts()/groupB.assignment_date.value_counts().sum())*100
```

### The phenomenon is present evenly in the two groups so it's safe to chose to only use first day assigned users as it wont produce bias.

```python
groupA = df1[((df1.abtest_group == 'A') & (df1.assignment_date == '2017-05-04'))]
groupB = df1[((df1.abtest_group == 'B') & (df1.assignment_date == '2017-05-04'))]
```

### Next I'll be producing some useful metrics

```python
print("""The percentage of users with convertions throughout the whole dataset is: {}%
The mean convertion delay of the whole dataset is: {}
The median convertion delay of the whole dataset is: {} 
The mode convertion delay of the whole dataset is: {}""".format(('%.3f'%((df1.conversion_date.count()/df1.playerid.count())*100)),df1.delay.mean(), df1.delay.median(), df1.delay.mode()))
```

```python
print("""The percentage of users with convertions in the A group is: {}%
The mean convertion delay of the A group is: {}
The median convertion delay of the A group is: {} 
The mode convertion delay of the A group is: {}""".format(('%.3f'%((groupA.conversion_date.count()/groupA.playerid.count())*100)),groupA.delay.mean(), groupA.delay.median(), groupA.delay.mode()))
```

```python
groupA.describe(include='all')
```

```python
print("""The percentage of users with convertions in the B group is: {}%
The mean convertion delay of the B group is: {}
The median convertion delay of the B group is: {} 
The mode convertion delay of the B group is: {}""".format(('%.3f'%((groupB.conversion_date.count()/groupB.playerid.count())*100)),groupB.delay.mean(), groupB.delay.median(), groupB.delay.mode()))
```
